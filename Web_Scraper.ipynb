```
import pandas as pd
import numpy as np
```

```
df1 = pd.read_csv('/content/drive/My Drive/DS/Data Masters Programs/data.csv')
df2 = pd.read_csv('/content/drive/My Drive/DS/Data Masters Programs/data (1).csv')
df3 = pd.read_csv('/content/drive/My Drive/DS/Data Masters Programs/data (2).csv')
```

```
df1.head()
```

```
url_list = df1['propertyName1_link'].tolist()
url_list2 = df2['propertyName1_link'].tolist()
url_list3 = df3['propertyName1_link'].tolist()
url_list.extend(url_list2)
url_list.extend(url_list3)
print(url_list)
```

```
import bs4
from urllib.request import urlopen, Request 
from bs4 import BeautifulSoup
import time
from random import randint
```

```
university_name =[]
program_name =[]
tags =[]
duration = []
tuition =[]
apply_date =[]
start_date =[]
description =[]
state = []
city = []
world_ranking = []
role =[]
disciplines =[]
delivered = []
i = 0
```

```
for url in url_list:
  print(i)
  hdr = {'User-Agent': 'Mozilla/5.0'}
  req = Request(url, headers=hdr)
  page = urlopen(req)
  soup = BeautifulSoup(page, 'html.parser')
  try:
    university_name_ = soup.h2(text = True)
    university_name.append(university_name_)
  except:
    university_name.append('NaN')
  try:
    program_name_ = soup.h1(text = True)
    program_name.append(program_name_)
  except:
    program_name.append('NaN')
  tag_container = soup.find('div', {'class' : 'DegreeTags'})  
  try:
    tags_ = tag_container.span(text = True)
    tags.append(tags_)
  except:
    tags.append('NaN')
  duration_container = soup.find('div', {'class': 'Title'})  
  try:
    duration_ = duration_container.span['data-duration']
    duration.append(duration_)
  except:
    duration.append('NaN')
  tuition_container = soup.find('div', {'class': 'TuitionFeeContainer'})  
  try:  
    tuition_ = tuition_container.span['data-amount']
    tuition.append(tuition_)
  except:
    tuition.append('NaN')
  date_container = soup.findAll('div', {'class': 'TimingContainer'})
  try: 
    applydate_ = date_container[0].time['datetime']
    apply_date.append(applydate_)
  except:
    apply_date.append('NaN')
  try:
    startdate_ = date_container[2].time['datetime']
    start_date.append(startdate_)
  except: 
    start_date.append('NaN')
  try: 
    description_ = soup.p(text = True)
    description.append(description_)
  except:
    description.append('NaN')    
  location_container = soup.find('div', {'class': 'js-locationInformationWrapper'})
  try: 
    state_ = location_container.span.find_next('span').find_next('span')(text=True)
    state.append(state_)
  except:
    state.append('Nan')
  try:
    city_ = location_container.span.a(text=True)
    city.append(city_)
  except:
    city.append('NaN')
  ranking_container = soup.find('div', {'class': 'Ranking'})  
  try:
    world_ranking_ = ranking_container.a.span.span(text=True)[0]
    world_ranking.append(world_ranking_)
  except:
    world_ranking.append('NaN')
  try:
    role_ = soup.find('div', {'class': 'FactListTitle js-durationFact'})(text =True)[0]
    role.append(role_)
  except:
    role.append('NaN')  
  try: 
    disciplines_ = soup.find('article', {'class': 'FactItem Disciplines'}).a['title']
    disciplines.append(disciplines_)
  except:
    disciplines.append('NaN')  
  try:  
    delivered_ = soup.findAll('article', {'class': 'FactItem'})[4].div(text=True)
    delivered.append(delivered_)
  except:
    delivered.append('NaN')
  time.sleep(randint(1,3))
  i += 1
```

```
df = pd.DataFrame(
    {'university_name': university_name,
     'program_name': program_name,
     'tags': tags,
     'duration': duration,
     'tuition': tuition,
     'apply_date': apply_date,
     'start_date': start_date,
     'description': description,
     'state': state,
     'city': city,
     'world_ranking': world_ranking,
     'role': role,
     'disciplines': disciplines,
     'delivered': delivered
    })
```
