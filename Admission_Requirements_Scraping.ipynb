```
import pandas as pd
import numpy as np
import bs4
from urllib.request import urlopen, Request 
from bs4 import BeautifulSoup
import time
from random import randint
```

```
df1 = pd.read_csv('/content/drive/My Drive/DS/Data Masters Programs/data.csv')
df2 = pd.read_csv('/content/drive/My Drive/DS/Data Masters Programs/data (1).csv')
df3 = pd.read_csv('/content/drive/My Drive/DS/Data Masters Programs/data (2).csv')
url_list = df1['propertyName1_link'].tolist()
url_list2 = df2['propertyName1_link'].tolist()
url_list3 = df3['propertyName1_link'].tolist()
url_list.extend(url_list2)
url_list.extend(url_list3)
print(url_list)
```

```
df = pd.read_csv('/content/drive/My Drive/DS/Data Masters Programs/masters_first_scraped.csv')
```

```
i = 0
mystring = '#content:requirement'
for s in url_list:
    url_list[i] = s+mystring
    i = i + 1
```

```
print(url_list[0])
```

```
gpa = []
gre = []
gmat = []
toefl = []
ielts = []
```

```
j = 0
for url in url_list:
  print(j)
  hdr = {'User-Agent': 'Mozilla/5.0'}
  req = Request(url, headers=hdr)
  page = urlopen(req)
  soup = BeautifulSoup(page, 'html.parser')
  try:
    gpa_container = soup.find('div', {'class' : 'CardContents GPACard js-CardGPA'})
    gpa_ = gpa_container.find_next('div').find_next('div').find_next('div')(text=True)[3]
    gpa.append(gpa_)
  except:
    gpa.append('NaN')
  gre_container = soup.find('div', {'class' : 'CardContents GRECard js-CardGRE'})
  try:
    gre_ = gre_container.find_next('div').find_next('div').find_next('div')(text=True)[1]
    gre.append(gre_)
  except:
    gre.append('NaN')
  gmat_container = soup.find('div', {'class' : 'CardContents GMATCard js-CardGMAT'})
  try:
    gmat_ = gmat_container.find_next('div').find_next('div').find_next('div')(text=True)[1] 
    gmat.append(gmat_)
  except:
    gmat.append('NaN')
  toefl_container = soup.find('div', {'class' : 'CardContents TOEFLCard js-CardTOEFL'})
  try:
    toefl_ = toefl_container.find_next().find_next('div').find_next('div').find_next('div')(text=True)[2]
    toefl.append(toefl_)
  except:
    toefl.append('NaN')
  ielts_container = soup.find('div', {'class' : 'CardContents IELTSCard js-CardIELTS'})
  try:
    ielts_ = ielts_container.find_next().find_next('div').find_next('div').find_next('div')(text=True)[2]
    ielts.append(ielts_)
  except:
    ielts.append('NaN')
  time.sleep(randint(1,3))
  j += 1
```
